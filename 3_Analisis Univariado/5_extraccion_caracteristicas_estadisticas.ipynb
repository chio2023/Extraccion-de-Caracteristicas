```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from funpymodeling.exploratory import freq_tbl
```

```python
# Asegúrate de que 'host_since' es de tipo fecha
data['host_since'] = pd.to_datetime(data['host_since'])

# Extraer el año de la columna 'host_since'
data['host_year'] = data['host_since'].dt.year

# Obtener las frecuencias de la columna 'host_year'
table16 = data['host_year'].value_counts().reset_index()  # 'value_counts' devuelve las frecuencias
table16.columns = ['host_year', 'frequency']  # Renombrar las columnas para mayor claridad

# Filtrar los valores con frecuencia mayor que 2
tableB16 = table16[table16['frequency'] > 1000]

# Ajustar el índice con la columna 'host_year'
Filtro16_index = tableB16.set_index('host_year')

# Mostrar el DataFrame resultante
Filtro16_index
```

```python
# Crear una gráfica de barras
Filtro16_index['frequency'].plot(kind='bar', figsize=(10,6), color='purple')

# Títulos y etiquetas
plt.title('Frecuencia de Hosts por Año', fontsize=16)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)

# Mostrar el gráfico
plt.show()
```
